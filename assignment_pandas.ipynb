## Проведите анализ данных при помощи Pandas выполнив поставленные задачи.
####

```python
import pandas as pd
```

```python
# загружаем датасет
data = pd.read_csv("./data/adult.data.csv")
data.head()
```

**1. Посчитайте, сколько мужчин и женщин (признак *sex*) представлено в этом датасете**

```python
gender_counts = data['sex'].value_counts()
print(gender_counts)
```

**2. Каков средний возраст мужчин (признак *age*) по всему датасету?**

```python
avg_age_men = data[data['sex'] == 'Male']['age'].mean()
print(f"Средний возраст мужчин: {avg_age_men:.1f} лет")
```

**3. Какова доля граждан Соединенных Штатов (признак *native-country*)?**

```python
us_citizens = data[data['native-country'] == 'United-States'].shape[0]
us_proportion = (us_citizens / data.shape[0]) * 100
print(f"Доля граждан США: {us_proportion:.2f}%")
```

**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

```python
# Те, кто получает >50K
high_income = data[data['salary'] == '>50K']['age']
print("Доход >50K:")
print(f"Средний возраст: {high_income.mean():.1f} лет")
print(f"Стандартное отклонение: {high_income.std():.1f} лет")

# Те, кто получает <=50K
low_income = data[data['salary'] == '<=50K']['age']
print("\nДоход <=50K:")
print(f"Средний возраст: {low_income.mean():.1f} лет")
print(f"Стандартное отклонение: {low_income.std():.1f} лет")
```

**6. Правда ли, что люди, которые получают больше 50k, имеют минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
higher_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']

high_income_edu = data[data['salary'] == '>50K']
no_higher_edu = high_income_edu[~high_income_edu['education'].isin(higher_education)]

if len(no_higher_edu):
    print(f"Нет, {len(no_higher_edu)} человек с доходом >50K не имеют высшего образования")
else:
    print("Да, все люди с доходом >50K имеют минимум высшее образование")
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

```python
age_stats = data.groupby(['race', 'sex'])['age'].describe()
print(age_stats)
```

```python
asian_men_max_age = data[(data['race'] == 'Asian-Pac-Islander') &
                          (data['sex'] == 'Male')]['age'].max()
print(f"Максимальный возраст мужчин расы Asian-Pac-Islander: {asian_men_max_age} лет")
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
men_data = data[data['sex'] == 'Male']

married_men = men_data[men_data['marital-status'].str.startswith('Married')]
single_men = men_data[~men_data['marital-status'].str.startswith('Married')]

married_high_income_pct = (married_men[married_men['salary'] == '>50K'].shape[0] / married_men.shape[0]) * 100
single_high_income_pct = (single_men[single_men['salary'] == '>50K'].shape[0] / single_men.shape[0]) * 100

print(f"Доля зарабатывающих >50K среди женатых мужчин: {married_high_income_pct:.2f}%")
print(f"Доля зарабатывающих >50K среди холостых мужчин: {single_high_income_pct:.2f}%")

if married_high_income_pct > single_high_income_pct:
    print("Вывод: среди женатых мужчин доля зарабатывающих >50K больше")
else:
    print("Вывод: среди холостых мужчин доля зарабатывающих >50K больше")
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours = data['hours-per-week'].max()
print(f"Максимальное число часов работы в неделю: {max_hours}")

people_max_hours = data[data['hours-per-week'] == max_hours]
print(f"Количество людей, работающих {max_hours} часов: {people_max_hours.shape[0]}")

high_income_max_hours_pct = (people_max_hours[people_max_hours['salary'] == '>50K'].shape[0] / people_max_hours.shape[0]) * 100
print(f"Процент зарабатывающих >50K среди них: {high_income_max_hours_pct:.2f}%")
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
avg_hours_by_country = data.groupby(['native-country', 'salary'])['hours-per-week'].mean().unstack().round(2)
print(avg_hours_by_country)
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
bins = [16, 36, 71, 101]
labels = ['young', 'adult', 'retiree']

data['AgeGroup'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
high_income_by_group = data[data['salary'] == '>50K'].groupby('AgeGroup').size()
print("Количество людей с доходом >50K в каждой возрастной группе:")
print(high_income_by_group)

total_by_group = data.groupby('AgeGroup').size()
pct_by_group = (high_income_by_group / total_by_group * 100).round(2)
print("\nДоля зарабатывающих >50K в каждой возрастной группе (%):")
print(pct_by_group)

best_group = pct_by_group.idxmax()
best_pct = pct_by_group.max()
print(f"\nВозрастная группа, в которой чаще зарабатывают >50K: {best_group}")
print(f"Доля зарабатывающих >50K в этой группе: {best_pct:.2f}%\n")
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
occupation_counts = data['occupation'].value_counts()
print("Количество людей в каждой группе занятости:")
print(occupation_counts)

def filter_func(group):
    return (group['age'].mean() <= 40) and (group['hours-per-week'].min() > 5)

filtered_occupations = data.groupby('occupation').filter(filter_func)

uniq_occupations = filtered_occupations['occupation'].unique()
print("\nПрофессии, удовлетворяющие условиям (средний возраст <= 40 и все работают >5 часов):")
for o in uniq_occupations:
    occ_data = data[data['occupation'] == o]
    print(f"  - {o}: средний возраст = {occ_data['age'].mean():.2f}, мин. часов = {occ_data['hours-per-week'].min()}")
```
