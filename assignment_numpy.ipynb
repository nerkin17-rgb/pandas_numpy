```python
import numpy as np
```

**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
arr = np.random.randint(0, 10, size=10)
print("Исходный массив:", arr)

arr[(arr > 3) & (arr < 8)] *= -1
print("Результат:", arr)
```

**2.** Заменить максимальный элемент случайного массива на 0

```python
arr = np.random.randint(0, 100, size=10)
print("Исходный массив:", arr)

arr[arr.argmax()] = 0
print("Результат:", arr)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
# Исходный массив
arr = np.array([[1, 2],
                [3, 4],
                [5, 6]])

n = arr.shape[0]

# В одну строку
result = np.hstack([np.repeat(arr, n, axis=0), np.tile(arr, (n, 1))])

print(result)
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
A = np.random.randint(1, 5, size=(8, 3))
B = np.random.randint(1, 5, size=(2, 2))

print("A:\n", A)
print("\nB:\n", B)

result = [i for i, row in enumerate(A)
          if all(set(b).issubset(set(row)) for b in B)]

print(f"\nИндексы строк: {result}")
print(f"Строки:\n{A[result]}")
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
M = np.random.randint(1, 3, size=(10, 3))
print("Исходная:\n", M)

result = M[np.array([len(set(row)) > 1 for row in M])]
print("\nРезультат:\n", result)
```

**6.** Дан двумерный массив. Удалить те строки, которые повторяются

```python
M = np.random.randint(1, 4, size=(8, 3))
print("Исходная матрица:\n", M)

result = np.unique(M, axis=0)
print("\nБез повторяющихся строк:\n", result)
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
X = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]

# Без numpy
product = 1
min_dim = min(len(X), len(X[0]))
for i in range(min_dim):
    if X[i][i] != 0:
        product *= X[i][i]
print(product)

# С numpy
diag = np.diag(X)
result = np.prod(diag[diag != 0])
print(result)
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
# Без numpy
x = [1, 2, 2, 4]
y = [4, 2, 1, 2]
result = sorted(x) == sorted(y)
print(result)

# С numpy
x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])
result = np.array_equal(np.sort(x), np.sort(y))
print(result)
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль.
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
# Без numpy
x = [6, 2, 0, 3, 0, 0, 5, 7, 0]

max_val = None
for i in range(1, len(x)):
    if x[i-1] == 0:
        if max_val is None or x[i] > max_val:
            max_val = x[i]
print(max_val)

# С numpy
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])

mask = np.roll(x, 1) == 0
mask[0] = False
result = np.max(x[mask]) if np.any(mask) else None
print(result)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# Без numpy
x = [2, 2, 2, 3, 3, 3, 5]

values = []
counts = []
current_val = x[0]
count = 1
for i in range(1, len(x)):
    if x[i] == current_val:
        count += 1
    else:
        values.append(current_val)
        counts.append(count)
        current_val = x[i]
        count = 1
values.append(current_val)
counts.append(count)
print(values, counts)



# С numpy
x = np.array([2, 2, 2, 3, 3, 3, 5])

change_indices = np.where(x[:-1] != x[1:])[0] + 1
values = x[np.concatenate(([0], change_indices))]
starts = np.concatenate(([0], change_indices))
ends = np.concatenate((change_indices, [len(x)]))
counts = ends - starts
print(values, counts)
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
# Без numpy
X = [[1, 2], [3, 4], [5, 6]]
Y = [[1, 1], [2, 2], [3, 3]]

import math
n, m = len(X), len(Y)
distances = [[0] * m for _ in range(n)]
for i in range(n):
    for j in range(m):
        sum_sq = 0
        for k in range(len(X[i])):
            sum_sq += (X[i][k] - Y[j][k]) ** 2
        distances[i][j] = math.sqrt(sum_sq)
for row in distances:
    print(row)


# С numpy
X = np.array([[1, 2], [3, 4], [5, 6]])
Y = np.array([[1, 1], [2, 2], [3, 3]])

X_sq = np.sum(X**2, axis=1).reshape(-1, 1)
Y_sq = np.sum(Y**2, axis=1).reshape(1, -1)
distances = np.sqrt(X_sq + Y_sq - 2 * np.dot(X, Y.T))
print(distances)
```
